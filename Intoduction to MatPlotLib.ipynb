## Matplotlib example workflow

```python
#0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig,ax=plt.subplots(figsize=(10,10)) #the numbers in the figsize brackets stands for (width, height)

#3. Plot data
ax.plot(x,y)

#4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

#5. Save and show (you save the whole figure)
fig.savefig("images/sample-plot.png")

plt.show()
```

## Two options for subplots

```python
# option 1 using a tuple
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,5))

ax1.plot(x, x/2);

ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());

```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10,8))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# option 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,8))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(exam_scores.keys(), exam_scores.values());
ax[1,1].hist(np.random.randn(1000));
```

# Pyplot vs OO method
When plotting something quickly, its okay to use the pyplot method
When plotting something more advanced use the OO method

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10,6));
```

```python
# OO method
fix, ax=plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

#ax.set_xlim([45, 100]);
```

```python
fig, ax=plt.subplots(figsize=(10, 6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["chol"],
                  c=over_50["target"])
```

```python
fig, ax=plt.subplots(figsize=(10, 6))

#plot the data
scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
#Subplots of chol, age, thalach: now we want to add a subplot to this
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 6),
                             sharex=True) #the sharex will amke both axes shares the same x axis

#add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");


#add data to ax1
scatter= ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"])

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax1.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

#add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
#set the style
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax=plt.subplots(figsize=(10, 6))

#plot the data
scatter= ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") #this codes changes the colur of spots

# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# how to constrain our plot to some values. customizing the x and y axis limitations

#Subplots of chol, age, thalach: now we want to add a subplot to this
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 6),
                             sharex=True) #the sharex will amke both axes shares the same x axis

#add data to ax0
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol");
ax0.set_xlim([50, 80]) #this changes the xlim of ax0

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax0.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");


#add data to ax1
scatter= ax1.scatter(x=over_50["age"],
                    y=over_50["thalach"],
                    c=over_50["target"],
                    cmap="winter")

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax1.set_xlim([50, 80]) #this adjusts/changes the x lim of ax1
ax1.set_ylim([60, 200]) #does the same but for y lim of ax1

#add a legend; with this we can tell what exactly the purple and yellow lines stands for
ax1.legend(*scatter.legend_elements(), title="Target");

#add a horizontal line
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--");

#add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
